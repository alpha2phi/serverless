```python
import io
import json
import base64
import requests
from PIL import Image
from IPython.display import Image as IPythonImage, display
```

```python
path = "../test_images/deeplab1.png"
img = Image.open(path)
img
```

```python
with open(path, "rb") as image_file:
    encoded_str = base64.b64encode(image_file.read()).decode()

url = "http://localhost:3000/dev/vision"

data = {
    "image": encoded_str
}
```

```python
%%time
response = requests.post(url, json=data)
display(IPythonImage(response.content))
```
